Recuperação de Informação de Artigos em HTML bruto via Google AdSense Tags
=======================

Fundamentos
----------
Alguns sites incluem no código html os comentários &lt;!-- google_ad_sense_start --> e &lt;!-- google_ad_sense_end --> para demarcar o conteúdo mais importante da página.

Outros fazem o contrário, atribuindo as tags &lt;!-- google_ad_sense_start(weight=ignore) --> e &lt;!-- google_ad_sense_end(weight=ignore) --> para demarcar conteúdo que não é central no assunto da página.

O objetivo é orientar o Google a analisar somente aquele trecho da página para gerar propagandas direcionadas no adSense.

Estratégia
-----------
Na suposição convincente de que as tags de AdSense são inseridas automaticamente pelo editor de texto usado no site, podemos acreditar que as tags estarão sempre corretamente posicionadas. Assim, a estratégia é uma simples rotina:

- Se há tags &lt;!-- google_ad_sense_start -->, selecionar apenas o conteúdo entre as tags &lt;!-- google_ad_sense_start --> e &lt;!-- google_ad_sense_end -->

- Caso contrário, se houver tags &lt;!-- google_ad_sense_start(weight=ignore) -->, selecionar apenas o conteúdo fora das tags &lt;!-- google_ad_sense_start(weight=ignore) --> e &lt;!-- google_ad_sense_end(weight=ignore) -->

Não são muitos os sites que apresentam as tags de AdSense, então na melhor das hipóteses a estratégia otimiza um subconjunto de artigos. Os testes preliminares apontam que o resultado é significativo (chega a descartar 90% de alguns documentos), indicando ser útil como etapa de pré-processamento antes de passar o documento a um HTML Stripper genérico.

Prévia dos Resultados
----------
- No banco de artigos analizados, 13% dos documentos eram suscetíveis à estratégia de selecionar.
- Algums sites utilizam variantes como google_ad_sense_start(name=default) ou google_ad_sense_start(name=s1,weight=3). Para melhores resultados deveríamos mimicar o comportamento de análise do adSense. Ou seja, processar os blocos de HTML conforme os argumentos da tag google_ad_sense_start().
- Aplicando a estratégia aos 65 artigos suscetíveis do banco, 90% do volume foi descartado.
- A estratégia é ótima para o caso de seleção positiva, por vezes eliminando 97~99% do volume dos documentos
- A estratégia é razoável para casos de seleção negativa, eliminando em torno de 70~90% do volume dos documentos

```python
# Get corpus archives from the dicrectory corpus/
import glob
html_files = glob.glob("raw_data/*")

print html_files[:3]

collection = list()
for path in html_files:
    f = open(path,'r')
    collection.append(f.read())
    f.close()


#collection
```

```python
pstart = "<!-- google_ad_section_start -->"
pstart2 = "<!-- google_ad_section_start(name=default) -->"
pend = "<!-- google_ad_section_end -->"
pend2 = "<!-- google_ad_section_end(name=default) -->"

nstart = "<!-- google_ad_section_start(weight=ignore) -->"
nend = "<!-- google_ad_section_start(weight=ignore) -->"

def AdStart(html):
    s1 = html.find(pstart)
    s2 = html.find(pstart2)
    if (s1!=-1 and s2!=-1):
        return min(s1,s2)
    else :
        return max(s1,s2)

def AdEnd(html):
    s1 = html.find(pend)
    s2 = html.find(pend2)
    if (s1!=-1 and s2!=-1):
        return min(s1,s2)
    else :
        return max(s1,s2)

subjects = [ (ID,html) for (ID,html) in enumerate(collection) if html.find("google_ad_section")!=-1]
print "{}/{} = {}% dos documentos é suscetível a análise via adSense.\n".format(len(subjects),len(collection),float(len(subjects))/len(collection)*100)

def PositiveSelection(html):
    start =  AdStart(html)
    end = AdEnd(html)
    if start==-1:
        return ""
    else:
        return (html[start:end] + PositiveSelection(html[end+1:])).replace(pstart,"").replace(pend,"")

def NegativeSelection(html):
    start =  html.find(nstart)
    end = html.find(nend)
    if start==-1:
        return html
    else:
        return NegativeSelection((html[:start]+html[end:]).replace(nstart,"").replace(nend,""))

def GoogleAdSenseTrick(html):
    if AdStart(html) != -1 :
        return PositiveSelection(html)
    else :
        return NegativeSelection(html)

total = 0
for (ID,html) in subjects:
    main = GoogleAdSenseTrick(html)
    pct = float(len(main))/len(html)*100
    total += pct
    print "{}/{} = {}% do documento foi considerado relevante.".format(len(main),len(html),pct)
    if len(main)<10 or pct>90:
        print html

print ""
print "Na média, apenas {}% do conteúdo dos arquivos é relevante.\n".format(total/len(subjects))
```
