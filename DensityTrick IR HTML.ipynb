Recuperação de Informação de HTML bruto de sites de notícias
====================
Fundamentos
-----------

O objetivo deste método é mimicar o algoritmo intuitivo e visual que aplicamos ao "passar o olho" num documento de HTML:

Para identificarmos uma região como de texto, descemos pelo documento até encontrar uma região densa em caracteres por linha, e verificamos se esse conteúdo é código estrutural (observando se grande parte do texto é tag HTML) ou é realmente texto. A partir daí, decidir se o texto é ou não o assunto da página requer leitura e análise contextual.

Exceto a etapa de análise contextual, o resto do algoritmo intuitivo pode ser codificado em termos de densidade de texto não-HTML por linhas de código no documento.

Para codificar o algoritmo de análise de densidade textual, podemos intuir as seguintes propriedades esperadas de um HTML que representa um artigo de notícia:

- O texto principal não está contido em tags HTML, scripts ou áreas de estilo

- O texto principal ocupa uma porção 'razoávelmente' contígua do documento. Isto é, raros blocos de propaganda, menus ou imagens entre parágrafos.

- Parágrafos textuais costumam ser formados por sentenças de mais de 5 palavras, gerando linhas longas.

Estratégia
-------------

Padronização de Formato e Limpeza de HTML:

- Padronizar estrutura via [HTML TIDY](http://www.w3.org/People/Raggett/tidy/) já desfazendo a identação

- Remover Scripts e Style (preservando espaços em branco) via HTMLParser

- Analizar o comprimento das linhas restantes no documento

Para decidir a área relevante do documento:

- Localizar regiões de pico na quantidade de caracteres por linha, caracterizando concentração de texto.

- Definir a região que apresenta muita informação, porém cercada de áreas de pouca informação

Observações
-------------

Nas aplicações atuais, os HTMLParsers da vida não fazem qualquer eliminação de lixo. Portanto, o objetivo é eliminar ALGUM lixo e evitar ao máximo eliminar qualquer TEXTO RELEVANTE. 

Confirmar o cumprimento desta condição requer verificação manual extensiva de exemplos, que não foi feita ainda.

Limpando documentos HTML
====================
Implementando handlers do HTMLParser para:

- Remover Scripts
- Remover Style
- Retornar o documento sem tags

Observações:

- Deixar espaço em branco onde era pra haver script/style. Mantem o tamanho e "esparsidade" do documento original.
- HTMLParser parece introduzir erro de codificação. Caracteres acentuados desaparecem. Provavelmente contornável.

Possibilidade:

- Talvez valha fazer uma gambiarra para aglutinar os textos, provavelmente é mais fácil detectar picos se o gráfico for aproximadamente contínuo.

```python
from HTMLParser import HTMLParser

class MyHTMLParser(HTMLParser):
    def __init__(self):
        HTMLParser.__init__(self)
        self.data = []
        self.tag = ""
    def handle_starttag(self, tag, attrs):
        self.tag = tag
    def handle_endtag(self, tag):
        pass
    def handle_data(self, d):
        if(self.tag not in ["style","script"]): # Mantém o conteúdo somente se não for style ou script
            self.data.append(d)
        else:
            blanked = [ '\n' for i in range(0, d.count('\n')) ] # Mas insere linhas em branco correspondentes
            self.data.append("".join(blanked))
    def parse(self, data):
        self.data = []
        self.tag = ""
        self.reset()
        self.feed(data)
        return "".join(self.data)

def cleanHTML(html):
    ## Aqui entra a possível gambiarra pra aglutinar texto. Usar com cautela por enquanto.
    html = html.replace("<div>&nbsp;</div>\n","") # "<div>&nbsp;</div>" é frequentemente usado como separador de parágrafos
    return MyHTMLParser().parse(html)

clean_docs = [cleanHTML(html) for html in raw_docs ]
```

Decidir área relevante
==================

Detecção de Picos
-------------

Calculamos uma medida de avaliação de picos (spikeness?) de uma região usando a diferença entre duas Moving Averages:

- Moving Averages de períodos mais longos tendem a suavizar a oscilação de uma curva. Ao subtrair uma MAVG de período médio por uma de periodo longo saberemos se uma região é de pico.

- As MAVGs adicionam a contribuição das linhas vizinhas ao cálculo da relevância de uma linha.

Isto é consistente com a intuição à priori de que um parágrafo de texto está próximo a outro parágrafo de texto, e lixo está próximo de espaço vazio.

Seleção de Região Relevante
----------------
Ao invés de definir um valor de corte para o spikeness de uma linha, a seleção deve levar em consideração a intuição à priori de que o texto ocupa uma porção contínua do documento.

Método:

- Buscamos o máximo global de spikeness, indicando concentração máxima de informação
- Buscamos os mínimos à direita e à esquerda da posição de máximo global, indicando um ponto de queda na quantidade de informação.

A degião definida pelo mínimo direito e mínimo esquerdo representa a área que contém brusco acréscimo+decréscimo de texto ao longo do documento. Esperamos que isso represente um texto grande em meio ao documento esparso.

```python
periodA = 20
periodB = 60

def relevanceGraph(html):
    clean = cleanHTML(html).split('\n')
    lenghts = [len(line) for line in clean]
    lenf = pandas.DataFrame(lenghts, columns=["Lenghts"])

    mavgA = pandas.rolling_mean(lenf,periodA).shift(-periodA/2)
    mavgA.columns=["mavgA"]
    mavgB = pandas.rolling_mean(lenf,periodB).shift(-periodB/2)
    mavgB.columns=["mavgB"]
    fig = mavgA.plot()
    mavgB.plot(ax=fig)
    
    spikeness = mavgA["mavgA"] - mavgB["mavgB"]
    lenf.plot(color='c')
    spikeness.plot(label="Spikeness")
    plt.legend()

def relevantSection(html):
    clean = cleanHTML(html).split('\n')
    lenghts = [len(line) for line in clean]
    lenf = pandas.DataFrame(lenghts, columns=["lenghts"])

    mavgA = pandas.rolling_mean(lenf,periodA).shift(-periodA/2)
    mavgA.columns=["mavgA"]
    mavgB = pandas.rolling_mean(lenf,periodB).shift(-periodB/2)
    mavgB.columns=["mavgB"]
    
    spikeness = mavgA["mavgA"] - mavgB["mavgB"]
    maximum = spikeness.idxmax()
    if maximum!=None:
        leftmin = (spikeness[:maximum]).idxmin()
        rightmin = (spikeness[maximum:]).idxmin()
        relevant = clean[leftmin:rightmin]
        
        return "\n".join(relevant)
    else:
        return None
```
